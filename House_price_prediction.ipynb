```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
dataset=pd.read_csv('boston.csv')
```

```python
print(dataset.head(5))
```

```python
print(dataset.shape)
```

```python
print(dataset.isnull().sum())
```

```python
print(dataset.describe())
```

```python
correlation=dataset.corr()
sns.heatmap(correlation,cbar=True,square=True,fmt='.2f',annot=True,annot_kws={'size':8},cmap='Blues')
```

```python
X=dataset.drop(['MEDV'],axis=1)
Y=dataset['MEDV']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=50)
print(X.shape,X_train.shape,X_test.shape)
print(Y.shape,Y_train.shape,Y_test.shape)
```

```python
model=XGBRegressor()
model.fit(X_train,Y_train)
train_data_prediction=model.predict(X_train)
test_data_prediction=model.predict(X_test)
```

```python
score1=metrics.r2_score(Y_train,train_data_prediction)
print('R squared error for training data :',score1)
score2=metrics.mean_absolute_error(Y_train,train_data_prediction)
print('Mean Absolute error for training data :',score2)
```

```python
score3=metrics.r2_score(Y_test,test_data_prediction)
print('R squared error for test data :',score3)
score4=metrics.mean_absolute_error(Y_test,test_data_prediction)
print('Mean Absolute error for test data :',score4)
```

```python
plt.scatter(Y_train,train_data_prediction)
plt.xlabel('Actual price')
plt.ylabel('Preddicted price')
plt.title('Actual Price vs Predicted Price')
plt.show()
```
